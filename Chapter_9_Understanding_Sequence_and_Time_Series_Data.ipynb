- **TÃ­nh nhiá»…u (Noise)**  

Sau Ä‘Ã³, tá»¥i mÃ¬nh sáº½ Ã¡p dá»¥ng cÃ¡c ká»¹ thuáº­t dá»± Ä‘oÃ¡n Ä‘á»ƒ kiá»ƒm tra xem hiá»‡u quáº£ nhÆ° tháº¿ nÃ o nha. ğŸ›ï¸ğŸ“Šâœ¨  

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_series(time, series, format="-", start=0, end=None):
  """
  HÃ m nÃ y dÃ¹ng Ä‘á»ƒ váº½ dá»¯ liá»‡u chuá»—i thá»i gian trong má»™t khung thá»i gian

  Tham sá»‘ Ä‘áº§u vÃ o:
  - time: khoáº£ng thá»i gian xáº£y ra
  - series: chuá»—i giÃ¡ trá»‹ tÆ°Æ¡ng á»©ng trong thá»i gian Ä‘Ã³
  - start: thá»i gian báº¯t Ä‘áº§u
  - end: thá»i gian káº¿t thÃºc
  """

  plt.plot(time[start:end], series[start:end], format)
  plt.xlabel("Time")
  plt.ylabel("Value")
  plt.grid(True)


def trend(time, slope=0):
  """
  HÃ m nÃ y giÃºp Ä‘iá»u hÆ°á»›ng xu hÆ°á»›ng cá»§a dá»¯ liá»‡u, giá»‘ng vá»›i phÆ°Æ¡ng trÃ¬nh báº­c 1

  Tham sá»‘ Ä‘áº§u vÃ o:
  - time: khoáº£ng thá»i gian xáº£y ra
  - slope: Ä‘á»™ dá»‘c cá»§a dá»¯ liá»‡u.
    + Náº¿u giÃ¡ trá»‹ dÆ°Æ¡ng thÃ¬ Ä‘Æ°á»ng tháº³ng hÆ°á»Ÿng lÃªn, ta cÃ³ xu hÆ°á»›ng Uptrend
    + Náº¿u giÃ¡ trá»‹ Ã¢m thÃ¬ chÃºng ta cÃ³ xu hÆ°á»›ng Downtrend

  Káº¿t quáº£ tráº£ vá»:
  - series: danh sÃ¡ch giÃ¡ trá»‹ theo xu hÆ°á»›ng thá»i gian
  """
  return slope * time


def seasonal_pattern(season_time):
  """
  HÃ m nÃ y dÃ¹ng Ä‘á»ƒ táº¡o ra máº«u dá»¯ liá»‡u trong 1 mÃ¹a.
  á» Ä‘Ã¢y mÃ¬nh chia lÃ m 2 pha tÃ­nh, logic tÃ­nh toÃ¡n nhÆ° sau nha:
  - Náº¿u thá»i gian mÃ¹a < 0.4, tiáº¿n hÃ nh tÃ­nh theo dao Ä‘á»™ng cosine: np.cos(season_time *2 *np.pi)
  - Náº¿u thá»i gian mÃ¹a > 0.4, tiáº¿n hÃ nh tÃ­nh theo hÃ m suy giáº£m mÅ©:  1 / np.exp(3 * season_time)

  Tham sá»‘ Ä‘áº§u vÃ o:
  - thá»i gian cá»§a mÃ¹a

  Káº¿t quáº£ tráº£ vá»:
  - res: káº¿t quáº£ tÃ­nh ra Ä‘Æ°á»£c tá»« logic tÃ­nh toÃ¡n náº±m trong khoáº£ng [-1, 1]
  """
  res = np.where(season_time<0.4,
                 np.cos(season_time * 2 *np.pi),
                 1 / np.exp(3 * season_time))
  return res

def seasonality(time, period, amplitude=1, phase=0):
  """
  HÃ m nÃ y Ä‘á»ƒ láº·p láº¡i, táº¡o nÃªn thá»i gian cho cÃ¡c mÃ¹a nha.
  Logic tÃ­nh toÃ¡n sáº½ nhÆ° sau:
  - time % period sáº½ ra giÃ¡ trá»‹ trong khoáº£ng [0, period - 1] -> táº¡o ra Ä‘Æ°á»£c cÃ¡c láº§n láº·p thá»i gian cá»§a mÃ¹a [0->period-1, 0->period-1,...]
  - / period sáº½ tiáº¿n hÃ nh chuáº©n hÃ³a láº¡i thá»i gian cá»§a mÃ¹a vá» khoáº£ng [0, 1]
  - Cháº¡y seasion time vÃ´ hÃ m seasonal_pattern Ä‘á»ƒ táº¡o ra chuá»—i giÃ¡ trá»‹ tÆ°Æ¡ng á»©ng vá»›i dÃ£y thá»i gian cá»§a cÃ¡c mÃ¹a, giÃ¡ trá»‹ náº±m trong khoáº£ng [-1, 1]
  - amplitude * seasonal_pattern, ta tiáº¿n hÃ nh nhÃ¢n biÃªn Ä‘á»™ dao Ä‘á»™ng vá»›i chuá»—i giÃ¡ trá»‹ thÃ¬ giÃ¡ trá»‹ cuá»‘i cÃ¹ng sáº½ trong khoáº£ng [-amplitude, amplitude]

  Tham sá»‘ Ä‘áº§u vÃ o:
  - time: khoáº£ng thá»i gian xáº£y ra
  - period: khoáº£ng thá»i gian cá»§a má»™t mÃ¹a
  - amplitude: biÃªn Ä‘á»™ dao Ä‘á»™ng tá»‘i Ä‘a cá»§a dá»¯ liá»‡u, vÃ­ dá»¥ mÃ¬nh Ä‘á»ƒ lÃ  15 thÃ¬ sáº½ dao Ä‘á»™ng trong khoáº£ng giÃ¡ trá»‹ tá»« -15 Ä‘áº¿n 15
  - phase: Ä‘á»™ lá»‡ch pha, vÃ­ dá»¥ náº¿u ta gÃ¡n 15, thay vÃ¬ mÃ¹a vá»¥ báº¯t Ä‘áº§u tá»« thá»i Ä‘iá»ƒm t = 0 thÃ¬ bá»‹ lá»‡ch Ä‘i 15 bÆ°á»›c thá»i gian tá»©c báº¯t Ä‘áº§u vÃ o t = 15

  Káº¿t quáº£ tráº£ vá»:
  - res: dá»¯ liá»‡u mang tÃ­nh chu ká»³, mÃ¹a vá»¥
  """
  season_time = ((time + phase) % period) / period
  res = amplitude * seasonal_pattern(season_time)
  return res


def noise(time, noise_level=1, seed=None):
  """
  HÃ m nÃ y Ä‘á»ƒ táº¡o tÃ­nh nhiá»…u cho dá»¯ liá»‡u
  Logic tÃ­nh toÃ¡n nhÆ° sau:
  - rnd.rand(len(time)) tráº£ vá» má»™t dÃ£y giÃ¡ trá»‹ cÃ³ Ä‘á»™ lá»›n trong khoáº£ng [0,1),
   sá»‘ lÆ°á»£ng giÃ¡ trá»‹ báº±ng sá»‘ bÆ°á»›c thá»i gian
  - Sau Ä‘Ã³ nhÃ¢n vá»›i noise_level sáº½ ra giÃ¡ trá»‹ trong khoáº£ng [0, noise_level)

  Tham sá»‘ Ä‘áº§u vÃ o:
  - time: khoáº£ng thá»i gian xáº£y ra.
  - noise_level: cáº¥p Ä‘á»™ nhiá»…u cá»§a dá»¯ liá»‡u
  - seed: náº¿u Ä‘á»ƒ giÃ¡ trá»‹ khÃ¡c None, thÃ¬ á»Ÿ cÃ¡c láº§n cháº¡y sau,
  khi giÃ¡ trá»‹ seed báº±ng nhau thÃ¬ giÃ¡ trá»‹ ngáº«u nhiÃªn táº¡o ra cÅ©ng báº±ng nhau

  Káº¿t quáº£ Ä‘áº§u ra:
  - res: dá»¯ liá»‡u mang tÃ­nh nhiáº¿u
  """
  rnd = np.random.RandomState(seed)
  res = rnd.rand(len(time)) * noise_level
  return res
```

á» dÆ°á»›i Ä‘Ã¢y má»i ngÆ°á»i cÃ³ thá»ƒ tá»± Ã½ thay Ä‘á»•i tham sao Ä‘á»ƒ táº¡o vÃ  quan sÃ¡t dá»¯ liá»‡u thÃªm nha.

```python
# Tiáº¿n hÃ nh táº¡o dá»¯ liá»‡u
time = np.arange(4 * 365 + 1, dtype='float32')
base_line = 10 # GiÃ¡ trá»‹ cÆ¡ sá»Ÿ ban Ä‘áº§u
amplitude = 15 # Cho biÃªn Ä‘á»™ dao Ä‘á»™ng
slope = 0.09 # Äá»™ dá»‘c cá»§a xu hÆ°á»›ng tÄƒng
noise_level = 15 # Äá»™ nhiá»…u
period = 365 # Khoáº£ng thá»i gian cá»§a má»™t mÃ¹a hay chu ká»³

# Tiáº¿n hÃ nh táº¡o dá»¯ liá»‡u bao gá»“m tÃ­nh xu hÆ°á»›ng, tÃ­nh chu ká»³ theo mÃ¹a vÃ  tÃ­nh nhiá»…u
series = (base_line + trend(time, slope)
          + seasonality(time, period, amplitude)
          + noise(time, noise_level, seed=42))
```

```python
# Biá»ƒu diá»…n dá»¯ liá»‡u
plt.figure(figsize=(8,5))
plot_series(time, series)
```

# Measuring Prediction Accuracy  
## (CÃ¡ch Ä‘o lÆ°á»ng Ä‘á»™ chÃ­nh xÃ¡c cá»§a dá»± Ä‘oÃ¡n)  

CÃ³ ráº¥t nhiá»u phÆ°Æ¡ng phÃ¡p Ä‘á»ƒ Ä‘o lÆ°á»ng Ä‘á»™ chÃ­nh xÃ¡c, nhÆ°ng táº¡m thá»i tá»¥i mÃ¬nh sáº½ táº­p trung vÃ o hai thang Ä‘o phá»• biáº¿n nháº¥t: **Mean Squared Error (MSE)** vÃ  **Mean Absolute Error (MAE)**. ğŸ“âœ¨  

### **1. Mean Squared Error (MSE)**  
- **Äá»‹nh nghÄ©a:** Trung bÃ¬nh cá»™ng bÃ¬nh phÆ°Æ¡ng sai sá»‘.  
- **CÃ´ng dá»¥ng:** TÃ­nh toÃ¡n Ä‘á»™ khÃ¡c biá»‡t trung bÃ¬nh giá»¯a giÃ¡ trá»‹ dá»± Ä‘oÃ¡n vÃ  giÃ¡ trá»‹ thá»±c táº¿, sá»­ dá»¥ng **bÃ¬nh phÆ°Æ¡ng** Ä‘á»ƒ khá»­ cÃ¡c giÃ¡ trá»‹ Ã¢m. Äiá»u nÃ y lÃ m tÄƒng trá»ng sá»‘ cho cÃ¡c sai sá»‘ lá»›n hÆ¡n, giÃºp phÃ¡t hiá»‡n cÃ¡c sai sá»‘ nghiÃªm trá»ng.  

### **2. Mean Absolute Error (MAE)**  
- **Äá»‹nh nghÄ©a:** Trung bÃ¬nh cá»™ng giÃ¡ trá»‹ tuyá»‡t Ä‘á»‘i sai sá»‘.  
- **CÃ´ng dá»¥ng:** TÃ­nh toÃ¡n Ä‘á»™ khÃ¡c biá»‡t trung bÃ¬nh giá»¯a giÃ¡ trá»‹ dá»± Ä‘oÃ¡n vÃ  giÃ¡ trá»‹ thá»±c táº¿, sá»­ dá»¥ng **giÃ¡ trá»‹ tuyá»‡t Ä‘á»‘i** Ä‘á»ƒ khá»­ cÃ¡c giÃ¡ trá»‹ Ã¢m, Ä‘áº£m báº£o tÃ­nh Ä‘Æ¡n giáº£n vÃ  dá»… hiá»ƒu hÆ¡n so vá»›i MSE.  

---

ChÃºng ta sáº½ tiáº¿n hÃ nh tÃ­nh toÃ¡n thá»­ cáº£ hai thang Ä‘o nÃ y trÃªn bá»™ dá»¯ liá»‡u Ä‘Ã£ táº¡o ra Ä‘á»ƒ xem sá»± khÃ¡c biá»‡t giá»¯a chÃºng nha. ğŸ§®ğŸ“Š

```python
import keras
from keras.metrics import mean_absolute_error, mean_squared_error
print("MAE: ", mean_absolute_error(series[split_time:], naive_forecast).numpy())
print("MSE: ", mean_squared_error(series[split_time:], naive_forecast).numpy())
```

# Less Naive: Using Moving Average for Prediction  
## (Sá»­ dá»¥ng trung bÃ¬nh Ä‘á»™ng Ä‘á»ƒ dá»± Ä‘oÃ¡n)  

BÃ¢y giá», tá»¥i mÃ¬nh sáº½ sá»­ dá»¥ng má»™t phÆ°Æ¡ng phÃ¡p dá»± Ä‘oÃ¡n phá»©c táº¡p hÆ¡n má»™t chÃºt nha. ğŸ“ˆâœ¨  
Ai tá»«ng tiáº¿p xÃºc vá»›i chá»©ng khoÃ¡n hay crypto cháº¯c khÃ´ng láº¡ gÃ¬ **giÃ¡ trá»‹ dá»± Ä‘oÃ¡n trung bÃ¬nh Ä‘á»™ng** hay **Ä‘Æ°á»ng trung bÃ¬nh Ä‘á»™ng** rá»“i ha. ğŸ’¹  

### **PhÆ°Æ¡ng phÃ¡p:**  
KhÃ¡c vá»›i dá»± Ä‘oÃ¡n Ä‘Æ¡n giáº£n trÆ°á»›c Ä‘Ã³, phÆ°Æ¡ng phÃ¡p nÃ y **nÃ¢ng cáº¥p** báº±ng cÃ¡ch tÃ­nh trung bÃ¬nh cá»™ng nhÃ³m dá»¯ liá»‡u trong má»™t khoáº£ng thá»i gian trÆ°á»›c Ä‘Ã³.  
- **CÃ´ng thá»©c:** GiÃ¡ trá»‹ dá»± Ä‘oÃ¡n táº¡i thá»i Ä‘iá»ƒm **t + 1** báº±ng **trung bÃ¬nh cá»™ng nhÃ³m dá»¯ liá»‡u trong khoáº£ng [t - n, t]**.  
- **VÃ­ dá»¥:** Náº¿u chá»n kÃ­ch thÆ°á»›c n lÃ  **30**, giÃ¡ trá»‹ dá»± Ä‘oÃ¡n trong tÆ°Æ¡ng lai sáº½ báº±ng **trung bÃ¬nh cá»™ng cá»§a 30 giÃ¡ trá»‹ trÆ°á»›c Ä‘Ã³**.  

### **GÃ³c nhÃ¬n ká»¹ thuáº­t:**  
Náº¿u má»i ngÆ°á»i Ä‘á»ƒ Ã½, nguyÃªn lÃ½ nÃ y ráº¥t giá»‘ng vá»›i:  
- **Cá»­a sá»• trÆ°á»£t (sliding window):** Trung bÃ¬nh cá»™ng giÃ¡ trá»‹ trong vÃ¹ng trÆ°á»£t kÃ­ch thÆ°á»›c **n**.  
- **Lá»›p Average Pooling** trong há»c sÃ¢u (deep learning): KÃ­ch thÆ°á»›c kernel **n** cÅ©ng táº¡o ra giÃ¡ trá»‹ trung bÃ¬nh tÆ°Æ¡ng tá»±.  

ÄÆ°á»ng trung bÃ¬nh Ä‘á»™ng giÃºp lÃ m mÆ°á»£t dá»¯ liá»‡u, giáº£m nhiá»…u vÃ  pháº£n Ã¡nh xu hÆ°á»›ng tá»‘t hÆ¡n so vá»›i cÃ¡c phÆ°Æ¡ng phÃ¡p cÆ¡ báº£n trÆ°á»›c Ä‘Ã³. ğŸ“Šâœ¨  

```python
def moving_average_forecast(series, window_size):
  """
  HÃ m nÃ y tiáº¿n hÃ nh tÃ­nh toÃ¡n giÃ¡ trá»‹ dá»± Ä‘oÃ¡n dá»±a trÃªn trung bÃ¬nh cá»™ng cá»­a sá»• trÆ°á»£t cÃ¡c giÃ¡ trá»‹ gáº§n nháº¥t
  Náº¿u kÃ­ch thÆ°á»›c cá»­a sá»• báº±ng 1 thÃ¬ phÆ°Æ¡ng phÃ¡p nÃ y láº¡i trá»Ÿ thÃ nh cÃ¡ch dá»± Ä‘oÃ¡n Ä‘Æ¡n giáº£n nhÆ° ban Ä‘áº§u nha.
  Logic tÃ­nh toÃ¡n: tiáº¿n hÃ nh tÃ­nh trung bÃ¬nh cá»™ng giÃ¡ trá»‹ cá»­a sá»• trÆ°á»£t tá»« bÆ°á»›c thá»i gian 0 Ä‘áº¿n t_max - window_size

  Tham sá»‘ Ä‘áº§u vÃ o:
  - series: giÃ¡ trá»‹ dá»¯ liá»‡u trong chuá»—i thá»i gian
  - window_size: kÃ­ch thÆ°á»›c cá»­a sá»• trÆ°á»£t

  Káº¿t quáº£ tráº£ vá»:
  - forecast: danh sÃ¡ch giÃ¡ trá»‹ dá»± Ä‘oÃ¡n
  """

  forecast = []
  for time in range(len(series) - window_size):
    forecast.append(series[time:time + window_size].mean())
  return np.array(forecast)
```

```python
window_size = 30
moving_avg = moving_average_forecast(series, window_size)[split_time - window_size:]
# Bá»Ÿi vÃ¬ khi tiáº¿n hÃ nh dá»± Ä‘oÃ¡n ta sáº½ pháº£i cáº¯t Ä‘i má»™t pháº§n dá»¯ liá»‡u bá»‹ máº¥t do window_size
# do Ä‘Ã³ index táº¡i vá»‹ trÃ­ split lÃ  split_time bá»‹ giáº£m Ä‘i 30
```

```python
plt.figure(figsize=(10,6))
plot_series(time_valid, x_valid)
plot_series(time_valid, moving_avg)
```

```python
# Tiáº¿n hÃ nh Ä‘Ã¡nh giÃ¡
print("MAE: ", mean_absolute_error(x_valid, moving_avg).numpy())
print("MSE: ", mean_squared_error(x_valid, moving_avg).numpy())
```

ÄÃ¢y lÃ  káº¿t quáº£ á»Ÿ láº§n cháº¡y cá»§a báº¡n:  

> **MAE:** 4.716281376001831  
> **MSE:** 34.42340844301733  

Uáº§y, váº­y lÃ  Ä‘á»™ chÃªnh lá»‡ch Ä‘Ã£ giáº£m Ä‘Ã¡ng ká»ƒ so vá»›i trÆ°á»›c rá»“i ha! ğŸ˜„  
Äiá»u nÃ y chá»©ng tá» phÆ°Æ¡ng phÃ¡p **trung bÃ¬nh Ä‘á»™ng (Moving Average)** Ä‘Ã£ hiá»‡u quáº£ hÆ¡n ráº¥t nhiá»u. ğŸ“Šâœ¨  

### **Háº¡n cháº¿ cá»§a phÆ°Æ¡ng phÃ¡p:**  
Tuy nhiÃªn, phÆ°Æ¡ng phÃ¡p nÃ y cÃ³ má»™t sá»‘ háº¡n cháº¿ nhÆ°:  
1. **Chá»‰ dá»±a trÃªn tÃ­nh tá»± tÆ°Æ¡ng quan (autocorrelation):**  
   - PhÆ°Æ¡ng phÃ¡p chá»‰ sá»­ dá»¥ng má»‘i quan há»‡ cá»§a **n giÃ¡ trá»‹ gáº§n nháº¥t** mÃ  khÃ´ng náº¯m báº¯t Ä‘Æ°á»£c cÃ¡c Ä‘áº·c tÃ­nh quan trá»ng khÃ¡c nhÆ°:  
     - **TÃ­nh xu hÆ°á»›ng (trend)**  
     - **TÃ­nh chu ká»³ theo mÃ¹a (seasonality)**  

2. **Hiá»‡u á»©ng trá»… (delay):**  
   - Khi biá»ƒu diá»…n giÃ¡ trá»‹ dá»± Ä‘oÃ¡n lÃªn biá»ƒu Ä‘á»“, ta cÃ³ thá»ƒ tháº¥y sá»± **biáº¿n Ä‘á»™ng bá»‹ cháº­m** hÆ¡n so vá»›i giÃ¡ trá»‹ thá»±c táº¿.
